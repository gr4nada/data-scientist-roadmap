## Linear Systems

Consider the following linear system of equations:
  
  $$ x + y = 5 $$
  $$ x-y = 1 $$

This system represents a pair of straight lines:

```python
x= np.linspace(-4, 8)
y1 = -x + 5
y2 = x - 1
plt.plot(x,y1)
plt.plot(x,y2)
```

Key questions:

1. Does the system have a solution?
2. Does the system have a approximate solution?

## Definition 1 :
### Linear Algebra is the study of linear system equations.

# Matrices and Determinants

A matrix is a rectangular array of numbers in rows and columns. For example,

$$A = \begin{bmatrix}5 & 4\\
6 & 7\end{bmatrix}$$

The matrix  has 2 rows and 2 columns; hence, it is a  matrix.

Systems of linear equations and linear transformations can be represented using matrices.

The determinant of a matrix is a number that is associated with a matrix. For example,

$$A = \begin{bmatrix}5 & 4\\
6 & 7\end{bmatrix} = 11$$

The determinant of a matrix encodes important information about the corresponding system or transformation.

## Modern Definition
### Linear algebra is the study of matrices and their determinants.
