```python
import numpy as np
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
```

## Linear Systems

Consider the following linear system of equations:
  
  $$ x + y = 5 $$
  $$ x-y = 1 $$

This system represents a pair of straight lines:

```python
x= np.linspace(-4, 8)
y1 = -x + 5
y2 = x - 1
plt.plot(x,y1)
plt.plot(x,y2)
```

Key questions:

1. Does the system have a solution?
2. Does the system have a approximate solution?

## Definition 1 :
### Linear Algebra is the study of linear system equations.


## Geometric Transformations

The following figure shows a reflection about the line $ x=y $ in the plane:

This is a linear geometric transformation, since the reflection of a line is another line.

Key questions:

1. Find an algebraic description of this transformation.

2. Are there any lines that are left unchanged by this transformation?

Definition 2: Linear algebra is the study of linear geometrical transformations.

```python
coord = np.array([[4,2], [5,9], [-3,5]])
coord.append(coord[0])
xs, ys = zip(*coord)

xs1 = *xs
ys1 = -1 * (*ys)
print(xs1)
plt.figure()
plt.plot(xs,ys) 
plt.plot(xs1,ys1) 
plt.show()
```
